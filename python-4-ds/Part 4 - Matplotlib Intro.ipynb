```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 10, 100)
y = np.sin(X)
```

### MATLAB Style

```python
plt.plot(X, y)
```

### Matplotlib Style

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(X, y)
```

# Standard Example

### MATLAB Style

```python
plt.figure(figsize=(8, 6))
plt.plot(X, y, "r--", linewidth=3, label="sinus")
plt.plot(X, np.cos(X), "b-.", linewidth=3, label="cosinus")
plt.xlim(0, 15)
plt.ylim(-2, 2)
plt.title("Trigonometri")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
```

### Matplotlib Style

```python
def test_function(*args, **kwargs):
    # positional arguments
    # keyword arguments
    print(args)
    print(kwargs)
    return

test_function(1, "a", 123, "hello", a="1", b="c", test="hello")
```

```python
fig = plt.figure(figsize=(8, 6))
ax = plt.axes()
ax.plot(X, y, "r--", linewidth=3, label="sinus")
ax.plot(X, np.cos(X), "c-.", linewidth=3, label="cosinus")
ax.set(xlim=(0,15), ylim=(-2,2), xlabel="x", ylabel="y", title="Trigonometri")
ax.legend()
```

    'b' = warna biru, solid line
    'ro' = warna merah, marker bulat
    'g-' = warna hijau, solid line
    '--' = warna default, garis dashed
    'k^:' = warna hitam, marker segitiga, garis dotted
    'c+-' = warna cyan, marker +, solid line

# Subplots

### MATLAB Style

```python
# plt.subplot(xyz) x-> jumlah baris, y -> jumlah colum, z -> posisi plot
plt.subplot(221)
plt.plot(X, y)

plt.subplot(223)
plt.plot(X, y)

plt.subplot(224)
plt.plot(X, np.cos(X))
```

### Matplotlib Style

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 8))

axes[0, 0].plot(X, y)
axes[0, 1].plot(X, np.cos(X))
axes[1, 1].plot(X, y)
```
