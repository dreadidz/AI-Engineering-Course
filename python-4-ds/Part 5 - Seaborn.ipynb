```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/linear.csv")
df.head()
```

```python
df1 = df[df.tipe == "data1"]
df2 = df[df.tipe == "data2"]
```

```python
plt.scatter(df1.X, df1.y, label="data1")
plt.scatter(df2.X, df2.y, label="data2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend();
```

# Statistical Summary

```python
df = pd.read_csv("data/cereal.csv", index_col="name")
df.head()
```

### Pairplot

```python
sns.pairplot(df.dropna(), hue="shelf", vars=["calories", "fat", "rating"])
```

# Untuk Categorical Data

### Persebaran

```python
sns.catplot("mfr", "rating", data=df)
```

```python
sns.boxplot("mfr", "rating", data=df)
```

### Frekuensi

```python
sns.countplot("mfr", data=df, hue="shelf")
```
