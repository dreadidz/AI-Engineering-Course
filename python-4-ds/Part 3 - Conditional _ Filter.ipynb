# New Column & Drop Column

```python
df.head()
```

```python
df["kolom_baru"] = 1
df.head()
```

```python
df["kolom_baru"] = np.random.rand(len(df))
df.head()
```

```python
df.drop(columns="kolom_baru", inplace=True)
df.head()
```

# Map, apply & transform

```python
df.shelf = df.shelf.map({1: "bawah", 2: "tengah", 3: "atas"})
df.head()
```

```python
def convert_to_mg(x):
    return x*1000
```

```python
df.fat = df.fat.apply(convert_to_mg)
df.head()
```

```python
df.vitamins = df.vitamins.apply(lambda x: x.lower())
df.head()
```

```python
df.rating = df.rating.transform(lambda x: x/x.sum())
df.head()
```

# One-hot Encoding / Dummy Variable

```python
df = pd.get_dummies(df, columns=["vitamins"])
df.head()
```

# Binning

```python
df["Grade"] = pd.cut(df.rating, bins=[0, 40, 70, 100], labels=["jelek", "lumayan", "enak"])
df.head()
```
