# Variable and Assignment

Syarat pembuatan variable:
1. Case insensitive
2. Tidak boleh menggunakan spasi dan simbol dan dimulai dari alphabet
3. Nama variable biasanya bahasa inggris
4. Nama variable yang lebih dari satu kata biasanya memiliki pola `CamelCase` atau `Underscore`. Pada bahasa pemrograman python `Underscore` lebih sering digunakan

Assignment adalah memasukan nilai kedalam variable

```python
number1 = 12
```

```python
number2 = 3
```

```python
total = number1 * number2
```

## Tuple

```python
animals = ["cat", "dog", "fish"]
animals[0] = "cow"
animals
```

```python
animals = ("dog", "cat", "cow") # animals = tuple(["cat", "dog", "fish"])
animals[0] = "cow"
animals
```

```python
animals.count("cow")
```

## Set

```python
animals = {"dog", "cat", "cow", "dog"} # animals = set(["dog", "cat", "cow", "dog"])
animals
```

```python
animals.add("fish")
animals
```
